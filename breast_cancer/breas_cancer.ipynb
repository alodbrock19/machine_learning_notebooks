```python
import pandas as pd 
```

```python
data = pd.read_csv("data.csv")
data
```

```python
data.info()
```

```python
# We drop column "Unnamed: 32" and "id"
data.drop(columns=["Unnamed: 32","id"],inplace=True)
data
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python

import math

# 1. Select only the numerical columns (exclude the diagnosis label itself)
# We drop 'diagnosis' so we don't try to plot the category against itself
features = data.select_dtypes(include=['float64', 'int64']).columns
features = [f for f in features if f != 'diagnosis']

# 2. Setup the grid (calculate how many rows/cols you need)
n_cols = 3  # Adjust this to change width
n_rows = math.ceil(len(features) / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
axes = axes.flatten()  # Flatten to make looping easy

# 3. Loop through every feature and give it a plot
for i, col in enumerate(features):
    sns.boxplot(data=data, x='diagnosis', y=col, ax=axes[i])
    axes[i].set_title(col)

# 4. Hide empty subplots if the number of features isn't a perfect multiple of n_cols
for i in range(len(features), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()

```

```python
data["diagnosis"] = data["diagnosis"].replace({"M":0, "B":1})
```

```python
from sklearn.model_selection import train_test_split

X = data.drop(columns=["diagnosis"])
y = data["diagnosis"]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled
```

```python
sns.heatmap(pd.DataFrame(X_scaled).corr(), annot=False, fmt=".2f", cmap="coolwarm")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test,y_pred)

print(f'model accuracy: {accuracy}')
```
