```python
from sklearn.model_selection import train_test_split

X = filtered_data.drop(columns="Yearly Amount Spent")
y = filtered_data["Yearly Amount Spent"]
X_train , X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
import statsmodels.api as sm
model = sm.OLS(y_train, sm.add_constant(X_train))
results = model.fit()
results.summary()
```

```python
# we drop "Time on Website" (no significant)
filtered_data = filtered_data.drop(columns="Time on Website")

X = filtered_data.drop(columns="Yearly Amount Spent")
y = filtered_data["Yearly Amount Spent"]
X_train , X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)

model2 = sm.OLS(y_train, sm.add_constant(X_train))
results2 = model2.fit()
results2.summary()
```
